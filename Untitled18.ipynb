```python
import numpy as np

def normalize_data(data):
    """
    Normalize the data using min-max scaling.
    """
    return (data - np.min(data)) / (np.max(data) - np.min(data))

def calculate_weighted_normalized_matrix(normalized_matrix, weights):
    """
    Calculate the weighted normalized decision matrix.
    """
    return normalized_matrix * weights

def calculate_ideal_and_negative_ideal(matrix, maximize):
    """
    Calculate the ideal and negative-ideal solutions.
    """
    ideal = np.max(matrix, axis=0) if maximize else np.min(matrix, axis=0)
    negative_ideal = np.min(matrix, axis=0) if maximize else np.max(matrix, axis=0)
    return ideal, negative_ideal

def calculate_euclidean_distance(matrix, ideal):
    """
    Calculate the Euclidean distance of each alternative from the ideal solution.
    """
    return np.sqrt(np.sum((matrix - ideal) ** 2, axis=1))

def calculate_topsis_score(matrix, ideal, negative_ideal):
    """
    Calculate the TOPSIS score (closeness to ideal) for each alternative.
    """
    d_pos = calculate_euclidean_distance(matrix, ideal)
    d_neg = calculate_euclidean_distance(matrix, negative_ideal)
    return d_neg / (d_pos + d_neg)

def topsis_recommendation(data, criteria_weights, maximize):
    """
    Perform TOPSIS analysis and provide smartphone recommendations.
    """
    # Step 1: Normalize the data
    normalized_data = normalize_data(data)

    # Step 2: Calculate the weighted normalized decision matrix
    weighted_normalized_matrix = calculate_weighted_normalized_matrix(normalized_data, criteria_weights)

    # Step 3: Calculate the ideal and negative-ideal solutions
    ideal, negative_ideal = calculate_ideal_and_negative_ideal(weighted_normalized_matrix, maximize)

    # Step 4: Calculate the TOPSIS score for each alternative
    topsis_scores = calculate_topsis_score(weighted_normalized_matrix, ideal, negative_ideal)

    # Step 5: Rank the alternatives based on TOPSIS scores
    ranked_indices = np.argsort(topsis_scores)[::-1]  # Sort in descending order

    return ranked_indices

# Sample data (rows represent smartphones, columns represent criteria)
data = np.array([
    [500, 5, 16, 12, 8],
    [600, 4, 32, 10, 9],
    [700, 4.5, 64, 14, 7],
    # Add more smartphone data here
])

# Sample weights (sum of weights should be 1)
criteria_weights = np.array([0.3, 0.2, 0.15, 0.2, 0.15])

# Set True for criteria to maximize (e.g., display quality), False for criteria to minimize (e.g., price)
maximize_criteria = np.array([False, True, True, False, True])

# Perform TOPSIS analysis and get the recommendation ranking
recommendation_ranking = topsis_recommendation(data, criteria_weights, maximize_criteria)

# Print the smartphone ranking based on TOPSIS scores
for rank, index in enumerate(recommendation_ranking, start=1):
    print(f"Rank {rank}: Smartphone {index+1}")
```
